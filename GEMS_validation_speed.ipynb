```python
# set up python environment
import pyreadr
import pandas as pd
import os
from os import listdir
import matplotlib.pyplot as plt
import seaborn as sns

os.chdir('/Users/xiaodanxu/Google Drive/GEMS/validation')

plt.style.use('ggplot')
```

```python
# load npmrds data for speed validation
speed_data_dir = 'speed/'
speed_data_source = 'NPMRDS'
output_plot_dir = 'plot/'
output_dir = 'output/'
list_of_speed_data_file = listdir(speed_data_dir + speed_data_source)
```

```python
# plot speed distribution
combined_speed_data = None
for file in list_of_speed_data_file:
    print('load observed speed data from ' + file)
    scenario_id = file.split('.')[2]
    npmrds_data = pyreadr.read_r(speed_data_dir + speed_data_source + '/' + file)
    npmrds_data_df = npmrds_data['all']
    npmrds_data_df['Scenario'] = scenario_id
    combined_speed_data = pd.concat([combined_speed_data, npmrds_data_df], sort = False)
    #print(npmrds_data_df.head(5))
    plt.figure(figsize=(8,6))
    sns.lineplot(x="hour", y="speed", hue="transgeo_microtype", palette = 'Spectral', data=npmrds_data_df)
    plt.ylabel('Observed Speed (mph)')
    plt.title(scenario_id)
    plt.savefig(output_plot_dir + speed_data_dir + 'hourly_speed_' + scenario_id + '.png', bbox_inches='tight', dpi = 300)
    plt.show()
    #break
```

```python
combined_speed_data = combined_speed_data.sort_values(by='transgeo_geotype') # help show subplot by alphabetic order

plt.figure(figsize=(12, 12))
sns.relplot(
    data = combined_speed_data, x = "hour", y = "speed", col = "transgeo_geotype", col_wrap = 2,
    hue="transgeo_microtype", kind="line", palette = 'Spectral'
)
plt.savefig(output_plot_dir + speed_data_dir + 'combined_NPMRDS_hourly_speed.png', bbox_inches='tight', dpi = 300)
plt.show()
```

```python
combined_speed_data.to_csv(output_dir + speed_data_dir + 'combined_NPMRDS_speed.csv',  index = False)
```
